### **Data unzipping**

```python
def unzip_gz_file(gz_file_path, output_file_path):
    """Unzips a .gz file to a specified output file path."""
    with gzip.open(gz_file_path, 'rb') as gz_file:
        with open(output_file_path, 'wb') as output_file:
            shutil.copyfileobj(gz_file, output_file)
    
    print(f"Decompressed {gz_file_path} to {output_file_path}")

def unzip_all_gz_files(directory):
    """Unzips all .gz files in a given directory and deletes the .gz files afterward."""
    for filename in os.listdir(directory):
        if filename.endswith('.gz'):
            gz_file_path = os.path.join(directory, filename)
            output_file_path = os.path.join(directory, filename[:-3])  # Remove .gz extension
            
            # Unzip the file
            unzip_gz_file(gz_file_path, output_file_path)
            
            # Delete the .gz file
            try:
                os.remove(gz_file_path)
                print(f"Deleted compressed file: {gz_file_path}")
            except OSError as e:
                print(f"Error deleting file {gz_file_path}: {e.strerror}")

```

```python
def get_all_files_in_folder(folder_path):
    # List all files in the specified folder
    file_names = os.listdir(folder_path)
    
    return file_names

# Example usage
folder_path = r"E:\BraTS2021_Training_Data"
patient_ids = get_all_files_in_folder(folder_path)
for patient_id in patient_ids[1:]:
    path = os.path.join(folder_path,patient_id)
    unzip_all_gz_files(path)
```

### **Dataset preparation**

```python
class BrainTumorDataset(Dataset):
    def __init__(self, root_dir, patient_ids, transform=None):
        """
        Args:
            root_dir (string): Path to the directory containing patient data.
            patient_ids (list): List of patient folder names (e.g., ['BraTS2021_00000', 'BraTS2021_00001']).
            transform (callable, optional): Optional transform to be applied on the data.
        """
        self.root_dir = root_dir
        self.patient_ids = patient_ids
        self.transform = transform

    def __len__(self):
        return len(self.patient_ids)

    def __getitem__(self, idx):
        patient_id = self.patient_ids[idx]
        patient_folder = os.path.join(self.root_dir, patient_id)

        # Load the four modalities (FLAIR, T1, T1ce, T2)
        flair_path = os.path.join(patient_folder, f"{patient_id}_flair.nii")
        t1_path = os.path.join(patient_folder, f"{patient_id}_t1.nii")
        t1ce_path = os.path.join(patient_folder, f"{patient_id}_t1ce.nii")
        t2_path = os.path.join(patient_folder, f"{patient_id}_t2.nii")
        seg_path = os.path.join(patient_folder, f"{patient_id}_seg.nii")

        # Load the images using nibabel and convert to numpy arrays
        flair = nib.load(flair_path).get_fdata()
        t1 = nib.load(t1_path).get_fdata()
        t1ce = nib.load(t1ce_path).get_fdata()
        t2 = nib.load(t2_path).get_fdata()
        seg = nib.load(seg_path).get_fdata()

        # Stack the modalities along the channel dimension
        image = torch.tensor([flair, t1, t1ce, t2], dtype=torch.float32)  # Shape: (4, D, H, W)
        seg = torch.tensor(seg, dtype=torch.long)  # Segmentation mask (D, H, W)

        # Normalize images (optional)
        image = (image - image.min()) / (image.max() - image.min())

        if self.transform:
            image, seg = self.transform(image, seg)

        return image, seg

# Example usage
root_dir = r"E:\BraTS2021_Training_Data"
dataset = BrainTumorDataset(root_dir, patient_ids[1:])

# Load one sample
image, seg = dataset[0]
print(f"Image shape: {image.shape}, Segmentation shape: {seg.shape}")
```

```python
def visualize_modalities_and_segmentation(image, seg):
    """
    Visualizes each modality at the middle slice index along with the corresponding segmentation.

    Parameters:
        image (torch.Tensor): The MRI image tensor of shape (4, D, H, W).
        seg (torch.Tensor): The segmentation mask tensor of shape (D, H, W).
    """
    # Convert the tensors to numpy arrays for visualization
    image_np = image.numpy()  # Shape: (4, D, H, W)
    seg_np = seg.numpy()      # Shape: (D, H, W)

    # Calculate the middle slice index
    middle_index = image_np.shape[1] // 2  # Middle slice

    # Create a figure for visualization
    fig, axes = plt.subplots(1, 5, figsize=(25, 5))

    modalities = ['FLAIR', 'T1', 'T1ce', 'T2', 'Segmentation']
    
    for i in range(4):
        # Plot each modality at the middle slice
        axes[i].imshow(image_np[i, middle_index, :, :], cmap='gray')
        axes[i].set_title(f'{modalities[i]} Slice {middle_index}')
        axes[i].axis('off')

    # Plot the segmentation mask at the middle slice
    axes[4].imshow(seg_np[middle_index, :, :], cmap='jet', alpha=0.5)  # Segmentation mask
    axes[4].set_title(f'Segmentation Slice {middle_index}')
    axes[4].axis('off')

    plt.tight_layout()
    plt.show()

image, seg = dataset[0]
print(f"Image shape: {image.shape}, Segmentation shape: {seg.shape}")

# Visualize each modality and the segmentation at the middle index
visualize_modalities_and_segmentation(image, seg)
```

**Segmentation Labels in Brain Tumor Imaging**

The segmentation masks in brain tumor imaging typically have several unique labels that correspond to different tissue types or tumor regions. Below are the unique labels found in the segmentation mask along with their descriptions:

| Label | Name                                   | Description                                                    |
|-------|----------------------------------------|----------------------------------------------------------------|
| 0     | **Background**                         | Represents the area outside the tumor and healthy brain tissue. |
| 1     | **Necrotic and Non-enhancing Tumor Core** | Represents the necrotic tissue and core of the tumor that does not enhance with contrast. |
| 2     | **Peritumoral Edema**                 | Represents the edema surrounding the tumor, which may appear as a region of swelling. |
| 4     | **Enhancing Tumor**                    | Represents the enhancing portion of the tumor that appears after the administration of contrast agent. |

**Note:** The labels may vary slightly depending on the specific dataset and preprocessing steps. Always refer to the dataset documentation for exact definitions.

```python
import os
import torch
import nibabel as nib  # For reading .nii files
from torch.utils.data import Dataset, DataLoader

# Function to classify tumor grade based on segmented images
def classify_tumor_grade(segmented_image):
    """
    Classifies a tumor as high-grade (1) or low-grade (0) based on the volume of tumor subregions.
    """
    LABEL_NECROTIC = 1
    LABEL_EDEMA = 2
    LABEL_ENHANCING = 4

    # Calculate volumes (number of voxels) for each subregion
    necrotic_volume = torch.sum(segmented_image == LABEL_NECROTIC).item()
    edema_volume = torch.sum(segmented_image == LABEL_EDEMA).item()
    enhancing_volume = torch.sum(segmented_image == LABEL_ENHANCING).item()

    # Total tumor volume
    total_tumor_volume = necrotic_volume + edema_volume + enhancing_volume

    # Avoid division by zero
    if total_tumor_volume == 0:
        return 0  # Default to low-grade if no tumor is present

    # Proportions
    enhancing_proportion = enhancing_volume / total_tumor_volume
    necrotic_proportion = necrotic_volume / total_tumor_volume

    # Thresholds for classification
    ENHANCING_THRESHOLD = 0.3
    NECROTIC_THRESHOLD = 0.2

    if enhancing_proportion > ENHANCING_THRESHOLD or necrotic_proportion > NECROTIC_THRESHOLD:
        return 1  # High-grade
    else:
        return 0  # Low-grade


# PyTorch Dataset class
class TumorClassificationDataset(Dataset):
    def __init__(self, segmented_image_paths):
        self.segmented_image_paths = segmented_image_paths

    def __len__(self):
        return len(self.segmented_image_paths)

    def __getitem__(self, idx):
        # Load the segmented image
        image_path = self.segmented_image_paths[idx]
        segmented_image = self.load_image(image_path)
        
        # Classify the tumor as high-grade or low-grade
        grade = classify_tumor_grade(segmented_image)
        
        return segmented_image, grade

    def load_image(self, image_path):
        # Load .nii file and convert it to a PyTorch tensor
        nii_image = nib.load(image_path)
        img_data = nii_image.get_fdata()
        img_tensor = torch.tensor(img_data, dtype=torch.float32)
        
        return img_tensor


# Example usage
if __name__ == "__main__":
    # Root directory containing the segmented images
    root_dir = "E:\\BraTS2021_Training_Data"
    
    # List of patient IDs and segmented image paths (you can extend this as per your data)
    # patient_ids = ['BraTS2021_00000', 'BraTS2021_00001', 'BraTS2021_00002']
    segmented_image_paths = []
    for id in patient_ids:
        path = os.path.join(root_dir, id, f"{id}_seg.nii")
        segmented_image_paths.append(path)

    # Create the dataset
    tumor_dataset = TumorClassificationDataset(segmented_image_paths)
    
    # Create a DataLoader for batching
    data_loader = DataLoader(tumor_dataset, batch_size=4, shuffle=True)
    
    # Iterate through the dataset
    for images, labels in data_loader:
        print(f"Batch of images: {images.shape}")
        print(f"Batch of labels: {labels}")
```

```python
def save_dataset(dataset, save_path):
    images = []
    labels = []

    for idx in range(len(dataset)):
        image, label = dataset[idx]
        images.append(image.unsqueeze(0))  # Add a batch dimension
        labels.append(label)

    # Stack all images into a single tensor
    images_tensor = torch.cat(images, dim=0)
    labels_tensor = torch.tensor(labels, dtype=torch.long)

    # Save the dataset to a file
    torch.save({'images': images_tensor, 'labels': labels_tensor}, save_path)
    print(f'Dataset saved to {save_path}')
```
